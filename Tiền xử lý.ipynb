```python
import pandas as pd
```

```python
churn = pd.read_csv("churn.csv")
churn.head()
```

```python
def ChuyenDoi():
    list = ['Account Length','Day Mins','Day Calls','Eve Mins','Eve Calls','Night Mins','Night Calls','Intl Mins','Intl Calls']
    for attribute in list:
        a = churn[attribute].quantile([.25, .5, .75])
        count = 0
        for key in churn[attribute]:
            if key < a[0.25]:
                churn.at[count,attribute]= 1
            elif key < a[0.75]:
                 churn.at[count,attribute]= 2
            else:
                churn.at[count,attribute]= 3
            count += 1
```

```python
ChuyenDoi()
churn.to_csv('churn_process.csv',index=False) #chuyển đổi các giá trị 4 thuộc tính thành các mức phân vị.
```

```python
def DeleteRowFalseInChurn():
    count = 0
    for key in churn['Churn?']:
        if key == 'False.':
            churn.drop(count,inplace= True)
        count += 1
```

```python
DeleteRowFalseInChurn()
churn.to_csv('churn_True.csv',index=False) #Xóa các dòng giá trị False.trong thuộc tính Churn.
```

```python
churn = pd.read_csv("churn.csv")
def DeleteRowTwoValueNo():
    ChuyenDoi()
    for index in range(churn.shape[0]):
        if churn['VMail Plan'][index] == 'no' and churn["Int'l Plan"][index] == 'no':
            churn.drop(index,inplace = True)
```

```python
DeleteRowTwoValueNo()
churn.to_csv('churn_plan.csv',index=False) #Xóa các dòng Int'l Plan = no và Vmail plan = no.
```
